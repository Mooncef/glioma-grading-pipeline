```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import sklearn
```

```python
# Import data
dataset = pd.read_csv("TCGA_GBM_LGG_Mutations_all.csv")
df = pd.DataFrame(dataset)
df.head()
```

```python
data = pd.read_excel("test.xlsx")
data_test = pd.DataFrame(data)
data_test.head()
```

```python
df_concatenated = pd.concat([df, data], axis=0, ignore_index=True)
df_concatenated.shape
```

```python
# Identify missing value

df_concatenated.replace("--", float("NaN"), inplace=True)
df_concatenated.replace("not reported", float("NaN"), inplace=True)
df_concatenated.isnull().sum()
```

```python
X = df_concatenated.drop('Grade', axis=1)
X = X.drop('sample_id', axis=1)
X = X.drop('sample id', axis=1)
X.head()
```

```python
Y = df['Grade']
Y.head()
```

```python
# Converte age in days

for i in range(X.shape[0]):
  if not pd.isna(X.iloc[i, 1]):
      parts = X.iloc[i, 1].split(' ')
      years = int(parts[0])
      if len(parts) == 4 :
        days = int(parts[2])
        totalDays = years * 365 + days
      else :
        totalDays = years * 365
      X.iloc[i, 1] = totalDays

X.head()
```

```python
X = X.values
Y = Y.values
```

```python
# Handling the missing values

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(X)
X = imputer.transform(X)
```

```python
for i in range (X.shape[0]):
  for j in range(X.shape[1]):
    if X[i][j]=='Male' or X[i][j]== 'MUTATED' :
      X[i][j] = 1
    if X[i][j]=='Female' or X[i][j]== 'NOT_MUTATED' :
      X[i][j] = 0
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [2])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X[0:5])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
print(Y)
```

```python
test = X[-16:]
test.shape
```

```python
X = X[:846]
X.shape
```

```python
# Splitting dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)

num_classes = 2
input_shape = X.shape[1]
```

```python
print(X_train.shape)
print(X_train[:2])
print(X_test.shape)
print(X_test[:2])
```

```python
print(y_train.shape)
print(y_train)
print(y_test.shape)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train[:, 5:6] = sta.fit_transform(X_train[:, 5:6])
X_test[:, 5:6] = sta.transform(X_test[:, 5:6])
```

```python
print("Train : ",X_train[:2])
print("Test : ",X_test[:2])
```

```python
import keras
from keras import layers
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')
X_test = X_train.astype('float32')
y_test = y_train.astype('float32')
test = test.astype('float32')
```

```python
import keras_tuner as kt

def model_builder(hp):
    input = layers.Input(shape=input_shape)
    x = layers.Dense(hp.Int('units_1', min_value=32, max_value=512, step=32), activation="relu")(input)
    x = layers.Dropout(hp.Float('dropout_1', min_value=0.2, max_value=0.5, step=0.1))(x)
    x = layers.Dense(hp.Int('units_2', min_value=32, max_value=512, step=32), activation="relu")(x)
    x = layers.Dropout(hp.Float('dropout_2', min_value=0.2, max_value=0.5, step=0.1))(x)
    output = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.models.Model(input, output)
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

tuner = kt.RandomSearch(
    model_builder,
    objective='val_loss',
    max_trials=3)
```

```python
tuner.search(X_train, y_train, epochs=10, validation_split=0.25)
best_model = tuner.get_best_models()[0]
```

```python
best_model.summary()
```

```python
test_loss, test_accuracy = best_model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```
